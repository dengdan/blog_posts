```python
import numpy as np;
import matplotlib.pyplot as plt;
%matplotlib inline
from mpl_toolkits.mplot3d import Axes3D

# 空间圆是一个球面与平面的交点.
res = 200;
## 先画球
phi = np.linspace(0, 1, res)*np.pi;
alpha = np.linspace(0, 2, 2 * res)*np.pi;
phi, alpha = np.meshgrid(phi, alpha);
R = 10;
x = R * np.sin(phi) * np.cos(alpha);
y = R * np.sin(phi) * np.sin(alpha);
z = R * np.cos(phi);

fig = plt.figure()
ax = Axes3D(fig)
ax.set_aspect('equal');
ax.plot_surface(x, y, z, color='white');

plt.show();
```
